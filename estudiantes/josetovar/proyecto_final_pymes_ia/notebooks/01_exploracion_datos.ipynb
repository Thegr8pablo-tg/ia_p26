# 01 Exploracion de datos

Notebook de apoyo para revisar las bases sinteticas raw.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..') if Path.cwd().name == 'notebooks' else Path('.')
emp = pd.read_csv(ROOT / 'data/raw/clientes_empresariales.csv')
per = pd.read_csv(ROOT / 'data/raw/clientes_personales.csv')
emp.head(), per.head()
```

```python
pd.concat([
    emp.assign(universo='empresarial'),
    per.assign(universo='personal')
]).groupby('universo')[['num_transacciones', 'dias_activos', 'gasto_total', 'ingresos_totales']].mean().round(2)
```
